```python
'''Take two 3D coordinates from the user. Find out the distance between these points
using following distance functions.'''
import math

def euclidean_distance(x1, y1, z1, x2, y2, z2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)

def manhattan_distance(x1, y1, z1, x2, y2, z2):
    return abs(x2 - x1) + abs(y2 - y1) + abs(z2 - z1)

def minkowski_distance(x1, y1, z1, x2, y2, z2, k):
    return (abs(x2 - x1)**k + abs(y2 - y1)**k + abs(z2 - z1)**k)**(1/k)

# Get user input for coordinates
x1 = float(input("Enter x1 coordinate: "))
y1 = float(input("Enter y1 coordinate: "))
z1 = float(input("Enter z1 coordinate: "))

x2 = float(input("Enter x2 coordinate: "))
y2 = float(input("Enter y2 coordinate: "))
z2 = float(input("Enter z2 coordinate: "))

k = 3

euclidean = euclidean_distance(x1, y1, z1, x2, y2, z2)
manhattan = manhattan_distance(x1, y1, z1, x2, y2, z2)
minkowski = minkowski_distance(x1, y1, z1, x2, y2, z2, k)

print("Euclidean Distance:", euclidean)
print("Manhattan Distance:", manhattan)
print("Minkowski Distance (k=3):", minkowski)
```

```python
'''Import the Iris dataset. Write a program to obtain the Euclidian Distance Matrix for
all the data samples in the feature space. Distance metric is a 2D array, where the
(i,j)th entry represents the distance between the ith and jth sample points in the feature
space.'''
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics.pairwise import euclidean_distances

# Load Iris dataset
iris = load_iris()
X = iris.data

# Compute Euclidean distance matrix
distance_matrix = euclidean_distances(X)

print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

class_0 = df[df['target'] == 0]
class_1 = df[df['target'] == 1]
new_dataset = pd.concat([class_0, class_1], ignore_index=True)

x_attribute = 'sepal length (cm)'
y_attribute = 'sepal width (cm)'


plt.figure(figsize=(8, 6))
plt.scatter(new_dataset[x_attribute], new_dataset[y_attribute], c=new_dataset['target'], cmap='viridis')
plt.xlabel(x_attribute)
plt.ylabel(y_attribute)
plt.title('Scatter Plot of Samples from Two Output Classes')
plt.colorbar(label='Output Class')
plt.grid(True)
plt.show()

m = -0.5
c = 3
separating_line_equation = f"y = {m}x + {c}"
print("Equation of the separating line:", separating_line_equation)
```
